```python
def remove_dashes(x):
    if pd.isnull(x):
        return x
    if ' - ' in x:
        return x.split(' - ')
    return [x]

def remove_commas(x):
    r = []
    if not isinstance(x, list) and pd.isnull(x):
        return x
    for i, y in enumerate(x):

        if ', ' in y:
             r += y.split(', ')
        else:
            r.append(y)
    return r
```

```python
ibd['MAPPED_GENE_extracted'] = ibd['MAPPED_GENE'].apply(remove_dashes)
ibd['MAPPED_GENE_extracted'] = ibd['MAPPED_GENE_extracted'].apply(remove_commas)
ibd[['MAPPED_GENE_extracted', 'MAPPED_GENE']].head()
```

```python
import requests
import re
url_server = 'https://rest.ensembl.org'
re_location = re.compile(r'(?P<chrom>\w+):(?P<start>\d+)-(?P<end>\d+)')


def extract_chrom_start_end(location):
    m = re_location.match(location)
    return m.groupdict()


def url_get(url):
#     print(url)
    r = requests.get(url, headers={ "Content-Type" : "application/json"})
    if not r.ok:
        r.raise_for_status()
        raise Exception()
    return r.json()


def parse_children(child):
    results = []
    if 'children' in child:
        for children in child['children']:
            results += parse_children(children)
    if 'sequence' in child:
        c = child['sequence']
        if 'name' in c:
            name = c['name']
            location = c['location']
            info = extract_chrom_start_end(location)
            assert len(c['id']) ==  1
            info['id'] = c['id'][0]['accession']
            info['name'] = c['name']
            results.append(info)
    return results


def get_region_info_for_gene(gene):
    """
    there is a chance this will return > 1 regions for 1 gene - process both in the pipeline
    """
    url_endpoint = '/genetree/member/symbol/homo_sapiens/'
    url_options = '?prune_species=homo_sapiens;content-type=application/json'
    url = ''.join([url_server, url_endpoint, gene, url_options])
    data = url_get(url)
    
    """
    try:
    
    except e:
        filter on name to include gene name ie "IL10RB-001"
    """
#     return data
    tree = data['tree']
    if 'sequence' in tree:
        location = tree['sequence']['location']
        info = extract_chrom_start_end(location)
        info['id'] = tree['id']['accession']
    else:
        info = parse_children(tree)
        info = [x for x in info if f'{gene}-' in x['name']]
        assert len(info) == 1
    if isinstance(info, list):
        return info[0]
    else:
        return info
    return
```

```python
genes = set(annotations.SYMBOL)
more_genes = ['BCL-X', 'CTSB', 'NLRX1', 'ASC', 'POP1', 'IPAF', 'NLRC4', 'CARD18', 'NLRP3', 'NLRP1', 'HSP90AA1', 'MEFV', 'TNFSF14', 'NOD1', 'CASP1', 'SUGT1', 'NLRP5', 'TNF', 'TNFSF4', 'CD40LG', 'CARD4', 'PSTPIP1', 'NOD2', 'ICEBERG', 'AIM2', 'CIITA', 'NLRP12', 'TRA1', 'TNFSF11', 'BCL2L1', 'HSP90AB1', 'NLRP9', 'NLRP6', 'TNFSF5', 'NAIP', 'OX40L', 'NLRC5', 'BCL2', 'NLRP4', 'PYCARD', 'PYDC1', 'ICE', 'IPAF', 'HSP90B1', 'CASP5', 'BIRC1', 'NCRP1', 'AIM', 'NLR', 'IL1B', 'IL18', 'NLRC1', 'NLRC2', 'NLRC3', 'NLRC4', 'NLRC5', 'PYD', 'ASC', 'NAIP1', 'NAIP2', 'NAIP3', 'NAIP4', 'NAIP5', 'NAIP6', 'NACHT', 'PYCARD']
genes.update(more_genes)
```

```python
def get_overlap(genes, df, window_radius=50000):
    no_region_data = []
    found = []
    not_found = []
    for gene in set(genes):
        try:
            info = get_region_info_for_gene(gene)
        except:
#             print(f'Error getting region info for {gene}')
            no_region_data.append(gene)
            continue
        start = float(info['start'])
        end = float(info['end'])
        if info['chrom'] == 'X':
            continue
        chrom = float(info['chrom'])
        i = chrom == df.CHR_ID
        i &= (start - window_radius) <= df.CHR_POS
        i &= (end + window_radius) >= df.CHR_POS
#         print(gene, chrom, start, end)
        n = sum(i)
        if n:
            found.append((gene, str(n)))
        else:
            not_found.append(gene)
    print('No region data found:')
    print(', '.join(no_region_data))
    
    print('\nNo overlap:')
    print(', '.join(not_found))
    
    print('\nFound:')
    print('\n'.join([' '.join(x) for x in found]))
```
